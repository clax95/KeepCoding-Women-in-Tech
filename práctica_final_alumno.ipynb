```python
# Importamos el dataset de los datos de los pokémon
df = pd.read_csv('dataset_final.csv')

# Mostrar los primeros elementos para ver la apariencia
df.head()
```

```python
'''
1. ¿Cuántos pokémon hay en el dataset?
'''
# Elegir el parámetro que indica el número de filas (elementos)
n = df.shape[0]
print(f'En el dataset hay {n} pokémon en total.')
```

```python
'''
2. ¿Cuántos pokémon hay de tipo Poison? (Type 1)
'''
# Filtrar los elementos que tengan 'Poison' como 'Type 1'
# Elegir el parámetro que indica el número de filas (elementos) de la selección
poison1 = df[df['Type 1'] == 'Poison']
n_poison1 = poison1.shape[0]
print(f'Hay {n_poison1} pokémon de tipo Poison (Type 1).')
```

```python
'''
3. ¿Cuántos tipos diferentes de pokémon hay? (Type 1)
'''
# Crear una lista con los distintos valores de 'Type 1' y ver su longitud
type = df['Type 1'].unique()
n_type = len(type)
print(f'Hay {n_type} tipos diferentes de pokémon.')
```

```python
'''
2. ¿Cuántos pokémon hay de cada tipo? (Type 1)
'''
# Agrupar los elementos según 'Type 1', contar el número que hay de cada uno y ordenarlos de mayor a menor
n_type1 = df.groupby('Type 1')['Name'].count().sort_values(ascending=False)
print(f'El número de pokémon de cada tipo es:\n {n_type1}')

'''
2.1 ¿Cuál es el tipo de pokémon con más pokémon? (Type 1)
'''
# Filtrar el elemento correspondiente al tipo con más pokémon
type1_max = n_type1[n_type1 == n_type1.max()]
print(f'El tipo con más pokémon es {type1_max.index[0]} con {type1_max.values[0]} pokémon.')
```

```python
'''
3. ¿Cuál es el pokémon más rápido?
'''
# Filtrar el elemento que tenga el valor más alto para 'Speed'
speed_max = df[df['Speed'] == df['Speed'].max()]
name_speed_max = speed_max['Name'].values[0]
speed_speed_max = speed_max['Speed'].values[0]
print(f'El pokémon más rápido es {name_speed_max} con velocidad {speed_speed_max}')
```

```python
'''
4. ¿Cuántos pokémon tienen una defensa superior a 100?
'''
# Filtrar los elementos que tengan el valor 'Defense' superior a 100
# Elegir el parámetro que indica el número de filas (elementos) de la selección
defense100 = df[df['Defense'] > 100]
n_defense100 = defense100.shape[0]
print(f'Hay {n_defense100} pokémon con defensa superior a 100.')
```

```python
'''
5. ¿Cuántos pokémon tienen una defensa superior a 100 y una velocidad superior a 100?
'''
# Filtrar los elementos que tengan los valores 'Defense' y 'Speed' superiores a 100
# Elegir el parámetro que indica el número de filas (elementos) de la selección
defense100 = df['Defense'] > 100
speed100 = df['Speed'] > 100
def_sp_100 = df[defense100 & speed100]
n_def_sp_100 = def_sp_100.shape[0]
print(f'Hay {n_def_sp_100} pokémon con defensa y velocidad superior a 100.')
```

```python
'''
7. Crea un nuevo dataset con los pokémon de tipo Water y Fire como Type 1.
'''
# Filtrar los elementos que tengan 'Water' o 'Fire' como 'Type 1'
water1 = df['Type 1'] == 'Water'
fire1 = df['Type 1'] == 'Fire'
df_water_fire = df[water1 | fire1]
df_water_fire
```

```python
'''
8. Crea un nuevo dataset con los pokémon Legendary.
'''
# Filtrar los elementos que sean Legendary
legendary = df['Legendary'] == True
df_legendary = df[legendary]
df_legendary
```

```python
'''
9. Calcula el promedio de stats de los pokémon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary.
'''
# Se filtran los campos que se quieren conocer y se hace la media de cada uno
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
# Pokémon 'Legendary'
df_legendary = df_legendary[stats]
stats_legendary = df_legendary.mean()
print(f"Los valores medios de los pokémon 'Legendary' son:\n {stats_legendary}")
# Pokémon no 'Legendary'
nolegendary = df['Legendary'] == False
df_nolegendary = df[nolegendary][stats]
stats_nolegendary = df_nolegendary.mean()
print(f"Los valores medios de los pokémon 'No Legendary' son:\n {stats_nolegendary}")
```

```python
'''
10. Crea un nuevo dataframe con el resultado del anterior ejercicio comparando ambos promedios.

ejemplo:
                HP  Attack  Defense  Sp. Atk  Sp. Def  Speed
Legendary       99      90       89       91       94     90
No Legendary    80      95       75       12       43     87

'''
# Se crea el nuevo dataframe a partir del diccionario con los valores medios de ambos tipos
d = { 'Legendary': stats_legendary.values, 'No Legendary': stats_nolegendary.values}
df_means = pd.DataFrame(d, index=stats)
df_means.transpose()
```

```python
'''
11. Añade una nueva columna ['Doble Tipo'] al dataframe inicial que inndique si el pokémon tiene dos tipos o no.
'''
# Cambiar los valores 'NaN' por un espacio '' para poder leerlos
# Ver los elementos que tienen algún valor de Type vacío y darle el valor a la nueva columna
df = df.fillna('')
type1 = df['Type 1'] == ''
type2 = df['Type 2'] == ''
df['Doble Tipo'] = ~(type1 | type2)
df
```

```python
from functools import reduce

'''
13. Dada una lista de artículos con sus precios. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']

# Crear una serie con todos los artículos y otra con la compra
ds_articulos = pd.Series(articulos)
ds_compra = ds_articulos[compra]

# A. Una función que calcule el precio total de la compra
def total(serie):
    total = reduce(lambda x, y: x + y, serie.values, 0)
    return print(f'El precio total de la compra es {total:.2f}€.')

# B. Una función que calcule el precio total de la compra con un descuento del 10% 
def total10(serie):
    total10 = reduce(lambda x, y : x + y * 0.9, serie.values, 0)
    return print(f'El precio total de la compra con un 10% de descuento es {total10:.2f}€.')

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def totalif(serie):
    totalif = reduce(lambda x, y : x + y * 0.9 if reduce(lambda x, y: x + y, serie.values, 0) > 100 else x + y, serie.values, 0)
    return print(f'El precio total de la compra con un 10% de descuento si supera los 100 € es {totalif:.2f}€.')

# D. Una función que calcule el precio total aplicando el IVA (21%)
def totaliva(serie):
    totaliva = reduce(lambda x, y : x + y * 1.21, serie.values, 0)
    return print(f'El precio total de la compra sumándole el 21% de IVA es {totaliva:.2f}€.')

# E.lista los artículos cuyo precio es superior a 20€
# Se crea un dataframe con todos los artículos
articulos_lista = list(articulos.keys())
precios_lista = list(articulos.values())
dic_articulos = { 'Articulos': articulos_lista, 'Precios': precios_lista}
df_articulos = pd.DataFrame(dic_articulos)
# Filtrar los articulos con un precio mayor a 20
def mayor20(data):
    mayor20 = data['Precios'] > 20
    data20 = data[mayor20]
    data20 = list(data20['Articulos'])
    return print(f'Los artículos con un precio mayor que 20€ son: {data20}.')

total(ds_compra)
total10(ds_compra)
totalif(ds_compra)
totaliva(ds_compra)
mayor20(df_articulos)
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]

# Crear un dataframe con los datos
df_alumnos = pd.DataFrame(alumnos, columns=['Nombre','Apellido', 'Curso', 'Notas'])

# Definir la función que filtra por curso y devuelve los datos
def alumnos_filtro(curso, data):
    curso = str(curso)
    cursox = data['Curso'] == curso
    notas_media = []
    for nota in data['Notas']: notas_media.append(reduce(lambda x, y: x + y, nota)/len(nota))
    data['Nota Media'] = notas_media
    return data[cursox][['Nombre', 'Apellido', 'Nota Media']]

alumnos_filtro(1, df_alumnos)
```
